## Top Spenders

* Run basic calculations to obtain the results in the table below


* Create a summary data frame to hold the results


* Sort the total purchase value column in descending order


* Optional: give the displayed data cleaner formatting


* Display a preview of the summary data frame



```python
# Retrieve SN, Item ID, Item Name and Price
Top_Spenders = purchase_data[["SN", "Item ID", "Price"]]

# Groupby Popular Items
Grouped_Top_Spenders_df = Top_Spenders.groupby(["SN"])

# Calculate Table Values
Purchase_Count = Grouped_Top_Spenders_df["Price"].count()
Average_Price = Grouped_Top_Spenders_df["Price"].mean()
Total_Purchase = Grouped_Top_Spenders_df["Price"].sum()

# Create Data Frame
Top_Spenders_Summary = pd.DataFrame({"Purchase Count": Purchase_Count, "Average Purchase Price": Average_Price,
                                                "Total Purchase Value": Total_Purchase})

# Sort Data Frame by Purchase Count
Sorted_Top_Spenders = Top_Spenders_Summary.sort_values("Total Purchase Value", ascending = False)

# Format and Print Data Frame 
pd.options.display.float_format = '${:,.2f}'.format
Sorted_Top_Spenders.head()
```

## Most Popular Items

* Retrieve the Item ID, Item Name, and Item Price columns


* Group by Item ID and Item Name. Perform calculations to obtain purchase count, average item price, and total purchase value


* Create a summary data frame to hold the results


* Sort the purchase count column in descending order


* Optional: give the displayed data cleaner formatting


* Display a preview of the summary data frame



```python
# Retrieve Item ID, Item Name and Price
Popular_Items = purchase_data[["Item ID", "Item Name", "Price"]]

# Groupby Popular Items
Grouped_Popular_Items_df = Popular_Items.groupby(["Item ID", "Item Name"])

# Calculate Table Values
Purchase_Count = Grouped_Popular_Items_df["Price"].count()
Average_Price = Grouped_Popular_Items_df["Price"].mean()
Total_Purchase = Grouped_Popular_Items_df["Price"].sum()

# Create Data Frame
Most_Popular_Summary = pd.DataFrame({"Purchase Count": Purchase_Count, "Item Price": Average_Price,
                                                "Total Purchase Value": Total_Purchase})

# Sort Data Frame by Purchase Count
Sorted_Most_Popular_Summary = Most_Popular_Summary.sort_values("Purchase Count", ascending = False)

# Format and Print Data Frame 
pd.options.display.float_format = '${:,.2f}'.format
Sorted_Most_Popular_Summary.head()
```
